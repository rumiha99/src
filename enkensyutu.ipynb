```python
import os
import cv2
import numpy as np

# フォルダのパス
folder_path = "/home/data/0917/mask"

# 結果を保存するファイル
txtfile = '/home/data/0917/circle_centers.txt'
excircle = '/home/data/0917/extracted_circles/'

def process_image(file_path, filename):
    # 画像を読み込む
    image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

    # 円を検出する
    circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, dp=1.2, minDist=50, param1=50, param2=30, minRadius=0, maxRadius=0)

    if circles is not None:
        circles = np.round(circles[0, :]).astype("int")
        
        circle = circles[0]
        (x, y, r) = circle
        
        mask = np.zeros_like(image)
        cv2.circle(mask, (x, y), r, 255, -1)
        
        # 円形部分を抽出
        extracted_circle = cv2.bitwise_and(image, image, mask=mask)    
        
        # 円の中心座標と半径をファイルに保存
        filename_without_mask = filename.replace("_mask", "")
        with open(txtfile, 'a') as f:
            f.write(f'{filename_without_mask},{x},{y}\n')
            
        # 描画した画像を保存
        cv2.imwrite(os.path.join(excircle, filename), extracted_circle)
    else:
        print(f"Image: {filename}, 円が検出されませんでした")
        filename_without_mask = filename.replace("_mask", "")
        with open(txtfile, 'a') as f:
            x = int(image.shape[1] / 2)
            y = int(image.shape[0] / 2)
            f.write(f'{filename_without_mask},{x},{y}\n')

# フォルダ内のすべてのファイルを一覧
for root, dirs, files in os.walk(folder_path):
    for filename in files:
        # ファイルのフルパスを作成
        file_path = os.path.join(root, filename)

        # ファイルが画像であることを確認（拡張子が.pngまたは.jpg.JPEGの場合）
        if file_path.lower().endswith(('.png', '.jpg', '.jpeg')):
            process_image(file_path, filename)
```
