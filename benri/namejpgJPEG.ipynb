```python
import os
import sys
import re
import shutil
```

```python
#.jpgならファイルの先頭にjpg,JPEGならファイルの先頭にJPEGを追加する
def rename_files_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(".jpg"):
            new_name = "jpg_" + filename
        elif filename.endswith(".JPEG"):
            new_name = "JPEG_" + filename
        else:
            continue  # .jpgでも.JPEGでもないファイルはスキップ

        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_name)
        os.rename(old_file, new_file)
        print(f"Renamed: {old_file} -> {new_file}")

# 使用例
folder_path = "/home/coco-annotator/datasets/seg_train/.thumbnail"  # 対象フォルダのパスに置き換えてください
rename_files_in_folder(folder_path)
```

```python
def revert_renamed_files_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.startswith("jpg_"):
            new_name = filename[4:]  # "jpg_"を削除
        elif filename.startswith("JPEG_"):
            new_name = filename[5:]  # "JPEG_"を削除
        else:
            continue  # "jpg_"でも"JPEG_"でもないファイルはスキップ

        old_file = os.path.join(folder_path, filename)
        new_file = os.path.join(folder_path, new_name)
        os.rename(old_file, new_file)
        print(f"Renamed: {old_file} -> {new_file}")

# 使用例
folder_path = "/home/coco-annotator/datasets/seg_train"  # 対象フォルダのパスに置き換えてください
revert_renamed_files_in_folder(folder_path)
```
